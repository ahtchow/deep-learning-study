## Perceptron Trick

In the last section you used your logic and your mathematical knowledge to create perceptrons for some of the most common logical operators. In real life, though, we can't be building these perceptrons ourselves. The idea is that we give them the result, and they build themselves. For this, here's a pretty neat trick that will help us.

The basis of the algorithm:

1. Take a random starting line

![p](Images/q10.png)

2. Repeat between measuring error and adjusting the line using math

Suppose we have

Line: 3 * x_1 + 4 * x_2 - 10 = 0

![p](Images/q13.png)

Is above or below line?

Below = Add

Above = Subtract

To get closer to the misclassified point simply add/subtract the point from the equation (w/ a bias of 1). We do not want to drastically change our line so use a ***Learning Rate of 0.1***


For the second example, where the line is described by 3x_1+ 4x_2 - 10 = 0, if the learning rate was set to 0.1, how many times would you have to apply the perceptron trick to move the line to a position where the blue point, at (1, 1), is correctly classified?

Ans: 10

# Perceptron Algorithm

1) Start with random weights: w_1 ... w_n , b

For every misclassified point:
           
2.1) If prediction = 0 (add) -> For i = 1...n , w_i + learning rate * x_i

2.2) If prediction = 1 (subtract) -> For i = 1...n , w_i - learning rate * x_i

Here is the algorithm below:

```python
def perceptronStep(X, y, W, b, learn_rate = 0.01):
    for i in range(len(X)):
        y_hat = prediction(X[i],W,b)
        if y[i]-y_hat == 1:
            W[0] += X[i][0]*learn_rate
            W[1] += X[i][1]*learn_rate
            b += learn_rate
        elif y[i]-y_hat == -1:
            W[0] -= X[i][0]*learn_rate
            W[1] -= X[i][1]*learn_rate
            b -= learn_rate
    return W, b

# Helper

def stepFunction(t):
    if t >= 0:
        return 1
    return 0

def prediction(X, W, b):
    return stepFunction((np.matmul(X,W)+b)[0])

# Full Training

def trainPerceptronAlgorithm(X, y, learn_rate = 0.01, num_epochs = 25):
    x_min, x_max = min(X.T[0]), max(X.T[0])
    y_min, y_max = min(X.T[1]), max(X.T[1])
    W = np.array(np.random.rand(2,1))
    b = np.random.rand(1)[0] + x_max
    # These are the solution lines that get plotted below.
    boundary_lines = []
    for i in range(num_epochs):
        # In each epoch, we apply the perceptron step.
        W, b = perceptronStep(X, y, W, b, learn_rate)
        boundary_lines.append((-W[0]/W[1], -b/W[1]))
    return boundary_lines
```

# Multi-Class Classification and Softmax

So far we have models that tell us Yes/No, whether discrete or continious. But what if we want our model to tell us if something is red/blue/green?

## Softmax

The softmax functions allow us to apply the yes/no model for multiple classes (i.e is it a duck/beaver/walrus) and calculate an overall score (percentage that add up to 1 for all classes).

![p](Images/q21.png)

Suppose we are given a score for each class, we must now combine all the scores from each class into an overall percentage for each class. Exponential function is useful since it can turn negative scores into positives. 

![p](Images/q22.png)

Here is the code below:

```python
import numpy as np

# Write a function that takes as input a list of numbers, and returns
# the list of values given by the softmax function.
def softmax(L):
    expL = np.exp(L)
    softmaxScore =[]
    for i in expL:
        softmaxScore.append(i/sum(expL))
    
    return softmaxScore
```

Rather than aiming for a high number with cross-entropy the lower the score the better.

![p](Images/q25.png)

Cross Entropy = -Summation(log(x)) , graph on left should have log(0.2) not log(0.7)


Below is the code for calculating cross-entropy:


```python
# Write a function that takes as input two lists Y, P,
# and returns the float corresponding to their cross-entropy.
def cross_entropy(Y, P):
    Y = np.float_(Y)
    P = np.float_(P)
    return -np.sum(Y * np.log(P) + (1 - Y) * np.log(1 - P))
    
```

Now, we're finally ready for one of the most popular and useful algorithms in Machine Learning, and the building block of all that constitutes Deep Learning. The Logistic Regression Algorithm. And it basically goes like this:

1. Take your data
2. Pick a random model
3. Calculate the error
4. Minimize the error, and obtain a better model

### Calculateing the error 

The error is the summation of cross-entropy divided by the number of datapoints

### Minimizing the error function

Use gradient descent or differentiate the error function. First derive the error function to find new weights and use them as new weights for the activation function. Error(W,b) -> Error(W',b') such that σ(W', b') can be used.

![p](Images/q30.png)

# Gradient Descent

## Gradient Calculation

![p](Images/q31.png)

We need to take the partial derivatives of all variables to find the gradient. The first thing to notice is that the sigmoid function has a really nice derivative. Namely,

σ'(x) = σ(x)(1 − σ(x))


To find the new sigmoid function:

1. Find Gradient ∇E of Sigmoid function y = σ(w1x1+ ... wnxn + b) ----> ∇E=( ∂/∂w1 E, ... , ∂/∂wn E , ∂/∂b )

2. Introduce a learning rate α = 0.1

3. Subtract the new weights and biases to the previous biases

4. Result: y = σ(W'x + b')

![p](Images/q32.png)

Overall Remember:

1. w1' = wi + α[- (y - y_hat) * xi ]
2. b' = b + α(y - y_hat)

# Gradient Descent Algorithm 

1. Start with random line

For every point:

2.1) For i = 1...n

    2.1.1) Update Wi' = Wi - (α) * ∂/∂wi =  wi - α(y_hat - y) * xi

    2.1.2) Update b' = b - (α) * ∂/∂wb = b - α(y_hat - y)
    
3) Repeat Until Error is Small

![p](Images/q33.png)



```python
# Helper Functions 

def plot_points(X, y):
    #Filter Rejected and Accepted
    admitted = X[np.argwhere(y==1)]
    rejected = X[np.argwhere(y==0)]
    #Make Accepted Red, Rejected Blue
    plt.scatter([s[0][0] for s in rejected], [s[0][1] for s in rejected], s = 25, color = 'blue', edgecolor = 'k')
    plt.scatter([s[0][0] for s in admitted], [s[0][1] for s in admitted], s = 25, color = 'red', edgecolor = 'k')

def display(m, b, color='g--'):
    #Plot the the regression line/curve
    plt.xlim(-0.05,1.05)
    plt.ylim(-0.05,1.05)
    x = np.arange(-10, 10, 0.1)
    plt.plot(x, m*x+b, color)
```

## Implementing the basic functions
Here is your turn to shine. Implement the following formulas, as explained in the text.
- Sigmoid activation function

$$\sigma(x) = \frac{1}{1+e^{-x}}$$

- Output (prediction) formula

$$\hat{y} = \sigma(w_1 x_1 + w_2 x_2 + b)$$

- Error function

$$Error(y, \hat{y}) = - y \log(\hat{y}) - (1-y) \log(1-\hat{y})$$

- The function that updates the weights

$$ w_i \longrightarrow w_i + \alpha (y - \hat{y}) x_i$$

$$ b \longrightarrow b + \alpha (y - \hat{y})$$

```python
# Gradient Descent Functions

# Activation (sigmoid) function
def sigmoid(x):
    return 1/(1+ np.exp(-x))

# Output (prediction) formula
def output_formula(features, weights, bias):
    # Return sigmoid for dot product of features and weights ,plus bias
    return sigmoid(np.dot(features, weights)+bias)

# Error (log-loss) formula
def error_formula(y, output):
    #Return the error for one point
    return (-y*np.log(output)-(1-y)*np.log(1-output))
    
# Gradient descent step
def update_weights(x, y, weights, bias, learnrate):
    #Return new weight and bias
    y_hat = output_formula(x, weights, bias)
    error = y - y_hat
    weights += learnrate * error * x
    bias += learnrate * error
    return weights,bias
```

# Training Function

This function will help us iterate the gradient descent algorithm through all the data, for a number of epochs. It will also plot the data, and some of the boundary lines obtained as we run the algorithm.

```python
# Make sure the same numbers
np.random.seed(44)

# Epochs - Number of iterations
epochs = 100
learnrate = 0.01

def train(features, targets, epochs, learnrate, graph_lines=False):

    errors = []
    n_records, n_features = features.shape
    last_loss = None
    
    # Generate Random Weights and Start with no bias
    weights = np.random.normal(scale=1 / n_features**.5, size=n_features)
    bias = 0
    
    # For each iteration perform Gradient Descent Algorithm
    for e in range(epochs):
        
        # Create a Matrix of 0, size of weights
        del_w = np.zeros(weights.shape)
        
        # zip return a tuple for each iteratable
        for x, y in zip(features, targets):
            output = output_formula(x, weights, bias)
            error = error_formula(y, output)
            weights, bias = update_weights(x, y, weights, bias, learnrate)
        
        # Printing out the log-loss error on the training set
        out = output_formula(features, weights, bias)
        loss = np.mean(error_formula(targets, out))
        errors.append(loss)
        
        if e % (epochs / 10) == 0:
            print("\n========== Epoch", e,"==========")
            
            if last_loss and last_loss < loss:
                print("Train loss: ", loss, "  WARNING - Loss Increasing")
            
            else:
                print("Train loss: ", loss)
            
            last_loss = loss
            
            predictions = out > 0.5
            accuracy = np.mean(predictions == targets)
            print("Accuracy: ", accuracy)
        
        # Display
        if graph_lines and e % (epochs / 100) == 0:
            display(-weights[0]/weights[1], -bias/weights[1])
    
    # Plotting the solution boundary
    plt.title("Solution boundary")
    display(-weights[0]/weights[1], -bias/weights[1], 'black')

    # Plotting the data
    plot_points(features, targets)
    plt.show()

    # Plotting the error
    plt.title("Error Plot")
    plt.xlabel('Number of epochs')
    plt.ylabel('Error')
    plt.plot(errors)
    plt.show()
```

## Time to train the algorithm!
When we run the function, we'll obtain the following:
- 10 updates with the current training loss and accuracy
- A plot of the data and some of the boundary lines obtained. The final one is in black. Notice how the lines get closer and closer to the best fit, as we go through more epochs.
- A plot of the error function. Notice how it decreases as we go through more epochs.

```python
# Read in Data
data = pd.read_csv('intro_to_neural_network.csv', header = None)
data_points = np.array(data[[0,1]])
score = np.array(data[2])
plot_points(data_points, score)


train(data_points, score, epochs, learnrate, True)
```

# Back Propagation

***Backpropagation***: Updating the weights to achieve better model.

Backpropagation consists of:
 1. Doing the feedfoward operator
 2. Comparing output with desired value
 3. Calculating the error
 4. Running Feedfoward backwards (Backpropagation) to spread the error to the weights.
 5. Update the weights to get a better model
 6. Continue until the model is good enough
 
 
 RECALL: For single layered perceptron the weights were update by taking -∇E Gradient to find new weights. E(W) -> E(W')
  
 What about mult-layered? Same. Here is a visualization
 
 ![p](Images/q40.png)
 
 As the neural net backpropogates, it asks which models have the higher cross-entropy. For those models it changes the weights signifigantly to improve the models.
 
 
 The math, how the weight updates: 
 ![p](Images/q42.png)
 
 ![p](Images/q41.png)

1. For each point calculate the score through foward propagation. Denoted as y_hat.

2. Use the error function, by finding the gradient of the Error function with respect to each weight , since each perceptron has the attribute y = σ( W*x +b)

3. Use the weights of the gradient to update the old weights.

Simply, W' = W - learning rate * ∂E/∂W . Where W is a matrix of weights

Deriving σ(x) = 1/(1+ exp(-x)) gives:

σ'(x) = σ(x) * (1−σ(x))

# Backpropagate the error
Write the error term. Remember that this is given by the equation
(𝑦−𝑦̂ )𝜎′(𝑥)

An error is the difference between the observed value and the true value (very often unobserved, generated by the DGP).

A error term is the difference between the observed value and the predicted value (by the model)..

```python
#Helper Function

# Activation (sigmoid) function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
def sigmoid_prime(x):
    return sigmoid(x) * (1-sigmoid(x))
def error_formula(y, output):
    return - y*np.log(output) - (1 - y) * np.log(1-output)
```

```python
#Write the error term formula
def error_term_formula(x, y, output):
    return (y-output)*sigmoid_prime(x)
```

```python
# Neural Network hyperparameters
epochs = 1000
learnrate = 0.5

# Training function
def train_nn(features, targets, epochs, learnrate):
    
    # Use to same seed to make debugging easier
    np.random.seed(42)
    
    # Base Data
    n_records, n_features = features.shape
    last_loss = None
    
    #Initialize Random Weights (Arbitrary scale, size)
    weights = np.random.normal(scale=1 / n_features**.5, size=n_features)
    
    for epoch in range(epochs):
        
        # Initiliaze Gradient for Weights
        del_w = np.zeros(weights.shape)
        
        for x, y in zip(features.values, targets):
            
            # Activation of the output unit : Probability in this case!
            #   Notice we multiply the inputs and the weights here 
            #   rather than storing h as a separate variable 
            output = sigmoid(np.dot(x, weights)) 

            # The error, the target minus the network output
            error = error_formula(y, output)

            # The error term
            error_term = error_term_formula(x, y, output) 

            # The gradient descent step, the error times the gradient times the inputs
            del_w += error_term * x

        #Update the previous weights by considering the average stepping    
        weights += learnrate * del_w / n_records

        # Printing out the mean square error on the training set
        if epoch % (epochs / 10) == 0:
            #For each iteration calculate the new score and measure error
            out = sigmoid(np.dot(features, weights))
            loss = np.mean((out - targets) ** 2)
            print("Epoch:", epoch)

            if last_loss and last_loss < loss:
                print("Train loss: ", loss, "  WARNING - Loss Increasing")
            else:
                print("Train loss: ", loss)

            last_loss = loss
            print("=========")
            
    print("Finished training!")
    return weights
```

```python
#Train

weights = train_nn(features, targets, epochs, learnrate)
```

# Calculate accuracy on test data

```python
# Calculate accuracy on test data
test_out = sigmoid(np.dot(features_test, weights))
predictions = test_out > 0.5
accuracy = np.mean(predictions == targets_test)
print("Prediction accuracy: {:.3f}".format(accuracy))
```
